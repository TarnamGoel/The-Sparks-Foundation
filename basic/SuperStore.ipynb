## Perform ‘Exploratory Data Analysis’ on dataset ‘SampleSuperstore’
### Try to find out the weak areas where you can work to make more profit.

```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the dataset using pandas
data=pd.read_csv('SampleSuperStore.csv')
```

```python
# Checking the top 5 values of the dataset
data.head()
```

```python
# Checking the bottom 5 values of the dataset
data.tail()
```

```python
# Get the description of the dataset
data.describe()
```

Here we can see, total sale was 22638.48 and average profit made was 28.656896

```python
#Discount and profit relation
data.plot(x='Discount',y='Profit',style='o')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

We can see, when discount is 0.4 or more than that, then we are selling in loss

# Category wise Sale and Profit

```python
# Category vs Sales Graph
data.plot(x='Category',y='Sales',style='o')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
# Category vs profit graph
data.plot(x='Category',y='Profit',style='o')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

```python
data['Category'].value_counts().plot.pie(figsize=(10,7),autopct='%.2f')
data['Category'].value_counts().describe
```

#### Office Supplies make the highest sale i.e 6026 (60.30%). 
#### Lowest Sale is made by technology i.e 1847 (18.48%)

# Sub-Category Analysis

```python
data['Sub-Category'].value_counts().plot.pie(figsize=(15,10),autopct='%.2f')
data['Sub-Category'].value_counts().describe
```

#### Binders and papers hold a good market. But Copiers, machines, supplies and fasteners are selling very less. Appliances also makes less profit. Hence we are more sure that technical products are not fitting good in the market.

# Segment wise sales

```python
data['Segment'].value_counts().plot.pie(autopct='%.f')
data['Segment'].value_counts().describe
```

#### As a B2C business, company is doing good. But the home and office supply of the products are very less. This may be a indication of less awareness of the brand in the market.
